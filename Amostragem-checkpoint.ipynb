```python
df.shape
```

```python
len(df) // 100
```

```python
def amostra_sistemica(dataset, amostra):
    intervalo = len(dataset) // amostra
    random.seed(1)
    inicio = random.randint(0, intervalo)
    faixa = np.arange(inicio, len(dataset), step=intervalo)
    amostra = dataset.iloc[faixa]
    return amostra
```

```python
df_amostra_aleatoria_sistemica = amostra_sistemica(df, 10)
df_amostra_aleatoria_sistemica.shape
```

```python
def amostragem_estratificada(dataset, percentual):
    split = StratifiedShuffleSplit(test_size=percentual, random_state=1)
    for _, y in split.split(dataset, dataset['income']):
        df_y = df.iloc[y]
        return df_y
```

# Amostragem de Reservatório

```python
def amostragem_reservatorio(dataset, amostra):
    stream = list(range(len(dataset)))
    tamanho = len(dataset)
    reservatorio = [stream[i] for i in range(amostra)]
    i = 0
    while i < tamanho:
        j = random.randrange(i + 1)
        if j < amostra:
            reservatorio[j] = stream[i]
        i += 1
    return dataset.iloc[reservatorio]
```

```python
df_amostragem_reservatorio = amostragem_reservatorio(df, 100)
df_amostragem_reservatorio.shape
```
